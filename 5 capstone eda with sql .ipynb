## Overview of the DataSet

SpaceX has gained worldwide attention for a series of historic milestones.

It is the only private company ever to return a spacecraft from low-earth orbit, which it first accomplished in December 2010.
SpaceX advertises Falcon 9 rocket launches on its website with a cost of 62 million dollars wheras other providers cost upward of 165 million dollars each, much of the savings is because Space X can reuse the first stage.

Therefore if we can determine if the first stage will land, we can determine the cost of a launch.

This information can be used if an alternate company wants to bid against SpaceX for a rocket launch.

This dataset includes a record for each payload carried during a SpaceX mission into outer space.

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
#query
selectQuery = "SELECT DISTINCT LAUNCH_SITE FROM SPACEX"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
#query
selectQuery = "SELECT * FROM SpaceX WHERE Launch_Site LIKE 'CCA%' LIMIT 5"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf

        
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
#query
selectQuery = "SELECT SUM(Payload_Mass) AS Total_PayloadMass FROM SpaceX WHERE Customer LIKE 'NASA (CRS)'"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
#query
selectQuery = "SELECT AVG(Payload_Mass) AS Avg_PayloadMass FROM SpaceX WHERE Booster_Version = 'F9 v1.1'"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
#query
selectQuery = "SELECT MIN(Date) AS FirstSuccessfull_landing_date FROM SpaceX WHERE Landing_Outcome LIKE 'Success (ground pad)'"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
#query
selectQuery = "SELECT Booster_Version FROM SpaceX WHERE Landing_Outcome = 'Success (drone ship)' AND Payload_Mass > 4000 AND Payload_Mass < 6000"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
#query
selectQuery = "select count(*),'success' FROM SpaceX WHERE Mission_Outcome like 'Success%' union select count(*),'failure' FROM SpaceX WHERE Mission_Outcome like 'Failure%';"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
#query
selectQuery = "SELECT Booster_Version, Payload_Mass FROM SpaceX WHERE Payload_Mass = (SELECT MAX(Payload_Mass) FROM SpaceX) ORDER BY Booster_Version"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
#query
selectQuery = "SELECT Booster_Version, Launch_Site, Landing_Outcome FROM SpaceX WHERE Landing_Outcome LIKE 'Failure (drone ship)' AND Date BETWEEN '2015-01-01' AND '2015-12-31'"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
#query
selectQuery = "SELECT Landing_Outcome, COUNT(Landing_Outcome) FROM SpaceX WHERE DATE BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY Landing_Outcome ORDER BY COUNT(Landing_Outcome) DESC"

#retrieve the query results into a pandas dataframe
pdf = pd.read_sql(selectQuery, pconn)

#print the dataframe
pdf
```
